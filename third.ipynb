## Advance operation with Buisness examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point
])

print("=== Zomato Sales Data Analysis ===")
print("\n shape of data:", sales_data.shape)

print("\n Sample data for 1st 3 restaurants:\n", sales_data[:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))  # sum along columns
yeary_total = np.sum(sales_data[:, 1:], axis=0)
print("Yearly total sales from 2021 to 2024:", yeary_total)

# minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)  # axis=1 for rows
print("Minimum sales per restaurant from 2021 to 2024:", min_sales)

# maximum sales per year 
max_sales = np.max(sales_data[:, 1:], axis=0)  # axis=0 for columns
print("Maximum sales per year from 2021 to 2024:", max_sales)

#avg sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per restaurant from 2021 to 2024:", avg_sales)
```

```python
cumsum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales per restaurant from 2021 to 2024:\n", cumsum_sales)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum_sales, axis=0), )
plt.title("Average Cumulative Sales from 2021 to 2024")
plt.xlabel("Years (2021 to 2024)")
plt.ylabel("Average Cumulative Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

print("addition:", vector1 + vector2)

print("\n multiplication:", vector1 * vector2)
print("\n dot product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n angle between vectors (in radians):", angle)
```
