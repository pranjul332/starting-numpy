## learn about numpy

```python
import numpy as np
```

Creating array from list

```python
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:", arr_2d)
```

### array vs list

```python
list1 = [1, 2, 3, 4, 5]
print("List:", list1*2)

arr = np.array([1, 2, 3, 4, 5])
print("Array multiplied by 2:", arr*2)
```

```python
import time
start_time = time.time()
list_comp = [i*2 for i in range(1000000)]
print("List comprehension time: %s seconds" % (time.time() - start_time))

start_time = time.time()
arr = np.arange(1000000) * 2
print("NumPy array operation time: %s seconds" % (time.time() - start_time))
```

### creating array from scratch 

```python
zeros = np.zeros((3, 4))
print("Zeros Array:\n", zeros)

ones = np.ones((2, 5))
print("Ones Array:\n", ones)

full = np.full((3, 3), 7)
print("Full Array:\n", full)

randome = np.random.random((2, 2))
print("Random Array:\n", randome)

seq = np.arange(10, 21, 2)
print("Sequence Array:", seq)
```
